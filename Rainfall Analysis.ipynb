```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\Ronit Kumar Purohit\\Downloads\\archive (8)\\Monthly_Rainfall_From_1901_to_2017.csv")
```

```python
# Replace "NA" values with NaN for easier handling
df.replace("NA", np.nan, inplace=True)
```

```python
df.head()
```

```python
# Convert numeric columns to float
df[df.columns[2:]] = df[df.columns[2:]].astype(float)
```

```python
# Iterate over each row
for i in range(len(df)):
    row = df.iloc[i]

    # Iterate over each month column
    for month in range(1, 13):
        # Get the column name for the current month
        col_name = df.columns[month + 1]

        # Check if the value is NaN
        if pd.isna(row[col_name]):
            # Get the year and state
            year = row["YEAR"]
            state = row["States/UTs"]

            # Find the last five years' data for the same month and state
            last_years_data = df[
                (df["YEAR"] >= (year - 5)) & (df["YEAR"] < year) & (df["States/UTs"] == state)
            ]

            # Calculate the mean for the same month from the last five years
            mean_value = last_years_data[col_name].mean()

            # Replace the NaN value with the calculated mean
            df.at[i, col_name] = mean_value
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

```python
# Sort the data by year and states/UTs
df.sort_values(['YEAR', 'States/UTs'], inplace=True)

# Calculate the rain deficit for each row
def calculate_rain_deficit(row):
    # Check if it is the first year for a state
    if row['YEAR'] == df['YEAR'].min():
        return 0.0  # No previous year data available for comparison
    else:
        # Find the rain from the previous year for the same state
        previous_year = row['YEAR'] - 1
        previous_rain = df.loc[(df['YEAR'] == previous_year) & (df['States/UTs'] == row['States/UTs']), 'ANNUAL'].values

        if len(previous_rain) > 0:
            previous_rain = previous_rain[0]
            rain_deficit = row['ANNUAL'] - previous_rain
            if rain_deficit < 0:
                return 0.0  # No deficit, as the current year's rain is greater than the previous year
            else:
                return rain_deficit
        else:
            return 0.0  # No previous year data available for comparison

# Apply the calculate_rain_deficit function to each row and create a new column 'Rain Deficit'
df['Rain Deficit'] = df.apply(calculate_rain_deficit, axis=1)

# Print the resulting DataFrame with the rain deficit
print(df[['YEAR', 'States/UTs', 'ANNUAL', 'Rain Deficit']])
```
